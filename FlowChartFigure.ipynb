```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define log posterior in PyTorch
def log_posterior(theta):
    t1, t2 = theta[0], theta[1]
    return -2.5 * (5*t1**2 + 5*t2**2 - 0.6*t1*t2) + 10 * torch.sin(3*t1) * torch.cos(2*t2)

# Gradient ascent setup
theta = torch.tensor([-1, -2], dtype=torch.float32, requires_grad=True)
lr = 0.001
steps = 5000
path = []

for _ in range(steps):
    lp = log_posterior(theta)
    lp.backward()
    with torch.no_grad():
        theta += lr * theta.grad
        path.append(theta.clone())
        theta.grad.zero_()
        

# Convert path to numpy arrays for plotting
path = torch.stack(path)
theta1_path = path[:, 0].numpy()
theta2_path = path[:, 1].numpy()
z_path = np.array([log_posterior(torch.tensor([t1, t2])).item() for t1, t2 in zip(theta1_path, theta2_path)])

# Grid for surface
theta1 = np.linspace(-3, 3, 5000)
theta2 = np.linspace(-3, 3, 5000)
T1, T2 = np.meshgrid(theta1, theta2)
Z = -2.5 * (5*T1**2 + 5*T2**2 - 0.6*T1*T2) + 10 * np.sin(3*T1) * np.cos(2*T2)

# Define the z-offset for projection
z_offset = np.min(Z) - 40
```

```python
# Plotting
fig = plt.figure(figsize=(10, 7),dpi=300)
ax = fig.add_subplot(111, projection='3d')


# # Contours
# ax.contour(T1, T2, Z, zdir='z', offset=z_offset, cmap='viridis', linewidths=1.0,levels=20,zorder=0)
# # Projected path
# ax.plot3D(theta1_path, theta2_path,z_offset, color='black', linewidth=2,zorder=1)

# Surface
ax.plot_surface(T1, T2, Z, cmap='viridis', edgecolor='none', alpha=0.96,zorder=0)

# Path on surface (white outline then black line)
ax.plot3D(theta1_path, theta2_path, z_path, color='white', linewidth=2, alpha=1.0, zorder=10)
ax.plot3D(theta1_path, theta2_path, z_path, color='black', linewidth=1, zorder=11)


# Vertical line from start point
ax.plot3D(
    [theta1_path[0], theta1_path[0]],
    [theta2_path[0], theta2_path[0]],
    [z_path[0], z_offset],
    color='black',
    linewidth=1,
    linestyle='--'
)

# Vertical line from end point
ax.plot3D(
    [theta1_path[-1], theta1_path[-1]],
    [theta2_path[-1], theta2_path[-1]],
    [z_path[-1], z_offset],
    color='black',
    linewidth=1,
    linestyle='--'
)

# Clean appearance
ax.set_axis_off()
ax.set_zlim(np.min(Z) - 6, np.max(Z))
plt.tight_layout()
plt.savefig("log_posterior_surface.png", dpi=300, transparent=True)
```
